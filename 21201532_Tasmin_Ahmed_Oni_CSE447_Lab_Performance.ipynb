Message: **birds_of_a_same_feather_flock_together**

Hide the message inside the green layer of the image and extract the message again to verify

```python
from PIL import Image

def hide_message(image_path, message):
    # Load the image from the given path
    img = Image.open(image_path)

    # Convert the message into binary format
    binary_message = ''.join(format(ord(i), '08b') for i in message)

    # Get the width and height of the image
    width, height = img.size

    # Check if the message fits in the image
    max_message_length = (width * height) // 8
    if len(binary_message) > max_message_length:
        raise ValueError('Message too long to be hidden in the given image')

    # Create a new image for storing the modified pixels
    new_img = Image.new(img.mode, img.size)
    message_index = 0

    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            red, green, blue = pixel

            # If the message has been fully hidden, copy the original pixel
            if message_index == len(binary_message):
                new_img.putpixel((x, y), pixel)
                continue

            # Modify the least significant bit of the green channel
            green_binary = format(green, '08b')
            green_binary = green_binary[:-1] + binary_message[message_index]  # Modify the LSB
            message_index += 1

            # Convert back to integer
            green = int(green_binary, 2)

            # Reconstruct the pixel with the modified green channel
            new_pixel = (red, green, blue)
            new_img.putpixel((x, y), new_pixel)

    # Save the new image with the hidden message
    new_img.save('hidden_image.png')

def extract_message(image_path):
    # Load the image from the given path
    img = Image.open(image_path)

    # Get the width and height of the image
    width, height = img.size

    binary_message = ''

    for y in range(height):
        for x in range(width):
            pixel = img.getpixel((x, y))
            red, green, blue = pixel

            # Extract the least significant bit of the green channel
            green_binary = format(green, '08b')
            binary_message += green_binary[-1]  # LSB of green

    # Convert binary message to text
    message = ''
    for i in range(0, len(binary_message), 8):
        message += chr(int(binary_message[i:i+8], 2))

    return message


# Hide a message inside the image at 'c/image.png'
message = "birds_of_a_same_feather_flock_together"
hide_message('bird.jpg', message)

# Extract the hidden message from the saved image
extracted_message = extract_message('hidden_image.png')

print("Extracted Message:", extracted_message)
```
